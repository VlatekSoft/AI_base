## **1. React + Material UI**

Material UI (MUI) — это библиотека компонентов React, которая позволяет быстро и легко создавать красивые интерфейсы для веб-приложений.

### Основные возможности Material UI:
1. **Готовые компоненты**: кнопки, текстовые поля, карточки, меню и другие элементы интерфейса.
2. **Простая настройка**: можно изменять стиль и поведение компонентов.
3. **Адаптивный дизайн**: компоненты подстраиваются под размеры экрана автоматически.

### Из чего состоит MUI:
- **Компоненты**: базовые элементы интерфейса (Button, TextField, Card и др.).
- **Темизация**: настройка цветов, шрифтов и стилей.
- **Сетки (Grid)**: для создания адаптивного дизайна.

### Шпаргалка Material-UI

#### Контейнеры и ширина

**maxWidth** — максимальная ширина, используется в Container.  
Значения:  
- "xs" — до 600px (смартфоны).  
- "sm" — до 960px (планшеты).  
- "md" — до 1280px (ноутбуки).  
- "lg" — до 1920px. (ноутбуки+).  
- "xl" — больше 1920px. (ноутбуки+).  

**flexShrink** — определяет, насколько элемент может сжаться, если в контейнере недостаточно места.
Значения:
- **`flexShrink: 1`** — элемент может сжаться, если необходимо, чтобы поместиться в контейнер.
- **`flexShrink: 0`** — элемент не будет сжиматься и останется в своём изначальном размере.
- **По умолчанию:** значение `flexShrink` равно `1`, что означает, что элементы могут сжаться, если это нужно.


**flexGrow** — определяет, насколько элемент может расширяться, чтобы заполнить доступное пространство в контейнере.
Значения:
- **`flexGrow: 1`** — элемент может расширяться, если есть доступное пространство.
- **`flexGrow: 0`** — элемент не будет расширяться и останется в своём изначальном размере.
- **По умолчанию:** значение `flexGrow` равно `0`, то есть элементы не будут расширяться, если не задано другое значение.


#### Позиционирование
**position** — используется в AppBar.  
Значения:  
- "static" — позиция по умолчанию, не фиксируется.  
- "fixed" — фиксируется к верхней части окна.  
- "absolute" — размещается относительно родителя.  
- "sticky" — прилипает при прокрутке.  
- "relative" — позиция относительно обычного потока.  

#### Кнопки
**variant** — используется в Button.  
Значения:  
- "contained" — залитая кнопка.  
- "outlined" — кнопка с контуром.  
- "text" — текстовая кнопка без обводки.  

**color** — задает цвет кнопки.  
Значения:  
- "primary" — основной цвет темы.  
- "secondary" — вторичный цвет темы.  
- "error" — цвет ошибки.  
- "info" — информационный цвет.  
- "success" — цвет успешного действия.  
- "warning" — цвет предупреждения.  
- "inherit" — унаследованный цвет от родительского элемента.
- color="action" - цвет для интерактивных элементов

#### Типографика
**variant** — используется в Typography.  
Значения:  
- "h1" — заголовок уровня 1.  
- "h2" — заголовок уровня 2.  
- "h3" — заголовок уровня 3.  
- "h4" — заголовок уровня 4.  
- "h5" — заголовок уровня 5.  
- "h6" — заголовок уровня 6.  
- "subtitle1" — подзаголовок уровня 1.  
- "subtitle2" — подзаголовок уровня 2.  
- "body1" — основной текст.  
- "body2" — текст поменьше.  
- "button" — текст кнопки.  

#### Сетка (в Material-UI сетка всегда делится по умолчанию на 12 колонок)
**container** — используется в Grid. Делает элемент контейнером сетки.  
**padding** — используется в Grid. Добавляет внутренний отступ внутри контейнера.
**item** — используется в Grid. Делает элемент частью сетки.  
**spacing** — задает отступы между элементами.  
Значения:  
- Любое число кратное 8px. Пример: spacing={3} → 24px.  

**xs, sm, md, lg, xl** — задают ширину элемента в колонках сетки.  
Пример:  
- xs={6} — занимает 6 колонок на маленьких экранах из 12 (смартфон).  
- sm={6} — занимает 6 колонок на средних экранах из 12 (планшет).  
- md={3} — занимает 3 колонки на больших экранах из 12 (ноутбук).  

#### Прогресс и уведомления
**severity** — используется в Alert.  
Значения:  
- "success" — успех.  
- "error" — ошибка.  
- "warning" — предупреждение.  
- "info" — информация.  

**open** — используется в Snackbar.  
- true — уведомление открыто.  
- false — уведомление закрыто.  

**autoHideDuration** — время отображения уведомления (в миллисекундах).  

**onClose** — событие, которое выполняется при закрытии уведомления.

```python
# @title Установка Django-сервера и React-сервера
!pip install django
!pip install django-cors-headers
!pip install opencv-python-headless
!pip install Pillow
!pip install numpy
!pip install ultralytics
!pip install pyngrok

import os
import re
import json
from django.http import JsonResponse
from django.views.decorators.csrf import csrf_exempt
import cv2
from ultralytics import YOLO
from pyngrok import ngrok
from google.colab import userdata

# Перейдем в папку проекта и создадим Django проект
!django-admin startproject myproject
%cd /content/myproject
!python manage.py startapp imageprocessor

# Создаем файл логов заранее
!touch /content/myproject/django_requests.log

# Настроим Ngrok для туннелирования
ngrok_key = userdata.get("NGROK_KEY")
ngrok.set_auth_token(ngrok_key)

# Установка React и Material-UI
!curl -fsSL https://deb.nodesource.com/setup_18.x | bash -
!apt-get install -y nodejs
!npx --yes create-react-app react-app
%cd react-app
!npm install axios
!npm install @mui/material @emotion/react @emotion/styled
!npm install @mui/icons-material

%cd /content/myproject/imageprocessor
!mkdir models

%cd /content/myproject/imageprocessor/models
!wget -nc https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8n.pt

!mkdir -p /content/myproject/static


def start_react_server():
    """Функция для запуска React-приложения на порту 3000 с использованием Ngrok."""
    # Остановка запущенных процессов
    os.system("pkill -f manage.py")
    os.system("pkill -f ngrok")

    # Запуск React-приложения в фоновом режиме
    os.system("nohup npm start > react_output.log 2>&1 &")

    try:
        # Настройка Ngrok для проброса порта 3000
        public_url2 = ngrok.connect(3000)
        print(f"React-приложение доступно по адресу:")
        return public_url2
    except Exception as e:
        print(f"Ошибка при подключении Ngrok: {e}")
        return None


def start_django_server():
    """Функция для запуска Django-сервера в фоновом режиме."""
    # Перемещаемся в директорию проекта
    os.chdir("/content/myproject")

    # Применяем миграции базы данных
    os.system("nohup python manage.py migrate > migrate.log 2>&1")

    # Запускаем Django-сервер в фоновом режиме
    os.system("nohup python manage.py runserver 0.0.0.0:8000 > django.log 2>&1 &")

    # Настраиваем Ngrok для проброса порта 8000
    public_url = ngrok.connect(8000)
    print(f"Django-приложение доступно по адресу: {public_url}")

    return public_url

def update_react_component_with_new_url(public_url):
    # Переходим в папку сервера react
    %cd /content/myproject/react-app

    # Извлечение URL с использованием регулярного выражения
    public_url_str = str(public_url)
    match = re.search(r'https://[^\s"]+', public_url_str)

    if match:
        link_r = match.group(0)
    else:
        print("URL не найден.")
        return

    # Создаем новый React-компонент с актуальным адресом
    text_ob_react = f"""
import React, {{ useState }} from "react";
import "./styles.css";
import {{ Grid, Box, Button, Typography, Card, CardContent, Divider, Snackbar, Alert }} from "@mui/material";
import UploadFileIcon from "@mui/icons-material/UploadFile";
import SendIcon from "@mui/icons-material/Send";

function App() {{
  const [selectedFile, setSelectedFile] = useState(null);
  const [isProcessing, setIsProcessing] = useState(false);
  const [result, setResult] = useState(null);
  const [imageWithBoxes, setImageWithBoxes] = useState(null);
  const [snackbarOpen, setSnackbarOpen] = useState(false);

  const handleFileChange = (event) => {{
    const file = event.target.files[0];
    setSelectedFile(file);
    setResult(null);
    setImageWithBoxes(null);
    if (file) {{
      setSnackbarOpen(true); // Показываем уведомление
    }}
  }};

  const handleSnackbarClose = () => {{
    setSnackbarOpen(false); // Закрываем уведомление
  }};

  const handleSubmit = async (e) => {{
    e.preventDefault();

    if (!selectedFile) {{
      alert("Пожалуйста, выберите изображение для загрузки");
      return;
    }}

    const formData = new FormData();
    formData.append("file", selectedFile);

    setIsProcessing(true);

    try {{
      const response = await fetch(`{link_r}/api/yolo/`, {{
        method: "POST",
        body: formData,
      }});

      if (!response.ok) {{
        throw new Error("Ошибка сервера");
      }}

      const data = await response.json();

      setResult(data.detections);

      const format = selectedFile.type.includes("png") ? "png" : "jpeg";
      setImageWithBoxes(`data:image/${{format}};base64,${{data.image_data}}`);
    }} catch (error) {{
      console.error("Ошибка при обработке изображения:", error);
      alert("Произошла ошибка при распознавании. Попробуйте снова.");
    }} finally {{
      setIsProcessing(false);
    }}
  }};

  return (
    <Box
      sx={{{{
        background: "linear-gradient(180deg, #e3f2fd, #bbdefb)",
        height: "100vh",
        padding: "20px",
      }}}}
    >
      <Grid container spacing={{2}} sx={{{{ height: "100%" }}}}>
        {{/* Левая панель */}}
        <Grid item xs={{4}} md={{4}}>
          <Card
            sx={{{{
              height: "100%",
              padding: "20px",
              display: "flex",
              flexDirection: "column",
              gap: "20px",
              justifyContent: "flex-start",
              backgroundColor: "#ffffff",
              borderRadius: "12px",
              boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
            }}}}
          >
            <CardContent>
              <Typography variant="h6" textAlign="center" gutterBottom>
                Управление
              </Typography>
              <Divider />
            </CardContent>
            <Button
              variant="contained"
              component="label"
              color="primary"
              fullWidth
              startIcon={{<UploadFileIcon />}}
            >
              Загрузить изображение
              <input type="file" hidden accept="image/*" onChange={{handleFileChange}} />
            </Button>
            <Button
              variant="contained"
              color="secondary"
              fullWidth
              onClick={{handleSubmit}}
              disabled={{isProcessing}}
              endIcon={{<SendIcon />}}
            >
              {{isProcessing ? "Обрабатываем..." : "Отправить на распознавание"}}
            </Button>
          </Card>
        </Grid>

        {{/* Правая панель */}}
        <Grid item xs={{8}} md={{8}}>
          <Grid container spacing={{2}} sx={{{{ height: "100%" }}}}>
            {{/* Верхняя зона (Изображение с рамками) */}}
            <Grid item xs={{12}} sx={{{{ maxHeight: "50%", overflowY: "auto" }}}}>
              <Card
                sx={{{{
                  padding: "20px",
                  boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
                  textAlign: "center",
                  backgroundColor: "#ffffff",
                  borderRadius: "12px",
                }}}}
              >
                <Typography variant="h5" sx={{{{ marginBottom: "20px", textAlign: "center" }}}}>
                  Изображение с рамками
                </Typography>
                {{imageWithBoxes ? (
                  <img
                    src={{imageWithBoxes}}
                    alt="Результат с рамками"
                    style={{{{ maxWidth: "100%", borderRadius: "12px" }}}}
                  />
                ) : (
                  <Typography variant="body1">Изображение отсутствует</Typography>
                )}}
              </Card>
            </Grid>

            {{/* Нижняя зона (JSON-данные) */}}
            <Grid item xs={{12}} sx={{{{ maxHeight: "50%", overflowY: "auto" }}}}>
              <Card
                sx={{{{
                  padding: "20px",
                  boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
                  backgroundColor: "#ffffff",
                  borderRadius: "12px",
                }}}}
              >
                <Typography variant="h5" sx={{{{ marginBottom: "20px" }}}}>
                  Результаты распознавания
                </Typography>
                {{result ? (
                  <pre style={{{{ whiteSpace: "pre-wrap", wordBreak: "break-word" }}}}>
                    {{JSON.stringify(result, null, 2)}}
                  </pre>
                ) : (
                  <Typography variant="body1">Нет данных для отображения</Typography>
                )}}
              </Card>
            </Grid>
          </Grid>
        </Grid>
      </Grid>
      <Snackbar
        open={{snackbarOpen}}
        autoHideDuration={{3000}}
        onClose={{handleSnackbarClose}}
        anchorOrigin={{{{ vertical: "bottom", horizontal: "center" }}}}
      >
        <Alert onClose={{handleSnackbarClose}} severity="success" sx={{{{ width: "100%" }}}}>
          Изображение успешно загружено!
        </Alert>
      </Snackbar>
    </Box>
  );
}}

export default App;
"""

    # Сохраняем обновленный файл App.js
    with open("src/App.js", "w") as file:
        file.write(text_ob_react)
    print("React компонент успешно обновлен с новым адресом.")



# Обновление файл settings.py с новыми настройками

# Путь к файлу settings.py
file_path = "/content/myproject/myproject/settings.py"

# Читаем содержимое файла settings.py
with open(file_path, "r") as file:
    content = file.read()

# Извлекаем значение SECRET_KEY
secret_key_match = re.search(r'SECRET_KEY\s*=\s*[\'"](.+?)[\'"]', content)
if secret_key_match:
    secret_key = secret_key_match.group(1)
else:
    raise ValueError("SECRET_KEY не найден в файле.")

# Новое содержимое settings.py
text_s = f"""
SECRET_KEY = '{secret_key}'

from pathlib import Path
import os

BASE_DIR = Path(__file__).resolve().parent.parent

DEBUG = True

ALLOWED_HOSTS = ['*']  # Разрешить запросы с любых хостов

INSTALLED_APPS = [
    'corsheaders',  # Добавить приложение для обработки CORS
    'django.contrib.admin',
    'django.contrib.auth',
    'django.contrib.contenttypes',
    'django.contrib.sessions',
    'django.contrib.messages',
    'django.contrib.staticfiles',
    'imageprocessor',  # Приложение для обработки изображений
]

MIDDLEWARE = [
    'corsheaders.middleware.CorsMiddleware',  # CORS обработчик
    'django.middleware.common.CommonMiddleware',
    'django.middleware.security.SecurityMiddleware',
    'django.contrib.sessions.middleware.SessionMiddleware',
    'django.middleware.csrf.CsrfViewMiddleware',
    'django.contrib.auth.middleware.AuthenticationMiddleware',
    'django.contrib.messages.middleware.MessageMiddleware',
    'django.middleware.clickjacking.XFrameOptionsMiddleware',
]

DATA_UPLOAD_MAX_MEMORY_SIZE = 10485760  # 10MB

CORS_ALLOW_ALL_ORIGINS = True  # Разрешить запросы с любых доменов
CORS_ALLOW_METHODS = ['GET', 'POST', 'OPTIONS']  # Разрешенные HTTP методы
CORS_ALLOW_HEADERS = ['content-type', 'authorization']  # Разрешенные заголовки

ROOT_URLCONF = 'myproject.urls'

TEMPLATES = [
    {{
        'BACKEND': 'django.template.backends.django.DjangoTemplates',
        'DIRS': [BASE_DIR / 'templates'],  # Папка для шаблонов, если нужна
        'APP_DIRS': True,
        'OPTIONS': {{
            'context_processors': [
                'django.template.context_processors.debug',
                'django.template.context_processors.request',  # Это важно для админки
                'django.contrib.auth.context_processors.auth',
                'django.contrib.messages.context_processors.messages',
            ],
        }},
    }},
]

WSGI_APPLICATION = 'myproject.wsgi.application'

DATABASES = {{
    'default': {{
        'ENGINE': 'django.db.backends.sqlite3',
        'NAME': BASE_DIR / 'db.sqlite3',
    }}
}}

AUTH_PASSWORD_VALIDATORS = [
    {{
        'NAME': 'django.contrib.auth.password_validation.UserAttributeSimilarityValidator',
    }},
    {{
        'NAME': 'django.contrib.auth.password_validation.MinimumLengthValidator',
    }},
    {{
        'NAME': 'django.contrib.auth.password_validation.CommonPasswordValidator',
    }},
    {{
        'NAME': 'django.contrib.auth.password_validation.NumericPasswordValidator',
    }},
]

LANGUAGE_CODE = 'en-us'

TIME_ZONE = 'UTC'

USE_I18N = True

USE_TZ = True

LOGGING = {{
    'version': 1,
    'disable_existing_loggers': True,  # Полностью отключить все стандартные логеры Django
    'formatters': {{
        'verbose': {{
            'format': '{{levelname}} {{asctime}} {{module}} {{message}}',
            'style': '{{',
        }},
    }},
    'handlers': {{
        'file': {{
            'level': 'INFO',  # Логируются только сообщения уровня INFO и выше
            'class': 'logging.FileHandler',
            'filename': os.path.join(BASE_DIR, 'django_requests.log'),
            'formatter': 'verbose',
        }},
    }},
    'loggers': {{
        'views': {{  # Логирование только из views.py
            'handlers': ['file'],
            'level': 'INFO',  # Логируются только сообщения уровня INFO и выше
            'propagate': False,  # Не передавать логи другим логгерам
        }},
    }},
}}

STATIC_URL = '/static/'
MEDIA_URL = '/media/'
MEDIA_ROOT = BASE_DIR / 'static/uploads'

# Если вы используете статику в разработке, укажите путь к папке
STATICFILES_DIRS = [
    BASE_DIR / "static",  # Папка с вашими статическими файлами
]

# Для продакшн-окружения
STATIC_ROOT = BASE_DIR / 'staticfiles'

DEFAULT_AUTO_FIELD = 'django.db.models.BigAutoField'
"""

# Перезаписываем файл settings.py с новыми настройками
with open(file_path, "w") as file:
    file.write(text_s)

# @title Обновление маршрутов для Django

# Переход в директорию проекта
%cd /content/myproject

# Создание папки для приложения imageprocessor, если её нет
os.makedirs('/content/myproject/imageprocessor', exist_ok=True)

# Создание файла urls.py в приложении imageprocessor
imageprocessor_urls_path = "/content/myproject/imageprocessor/urls.py"
with open(imageprocessor_urls_path, 'w') as f:
    f.write("""
from django.urls import path
from .views import process_image  # Импорт обработчика из views

urlpatterns = [
    path('yolo/', process_image, name='process_image'),  # Маршрут для обработки изображений
]
""")
print("Файл urls.py для приложения imageprocessor создан или обновлён.")

# Обновление основного файла urls.py проекта
project_urls_path = "/content/myproject/myproject/urls.py"
urls_content = """
from django.contrib import admin
from django.urls import path, include

urlpatterns = [
    path('api/', include('imageprocessor.urls')),  # Подключаем маршруты из imageprocessor
    path('admin/', admin.site.urls),
]
"""

with open(project_urls_path, "w") as file:
    file.write(urls_content)

print("Файл urls.py для проекта обновлён.")

# Создание файла views.py в приложении imageprocessor, если его нет
imageprocessor_views_path = "/content/myproject/imageprocessor/views.py"
if not os.path.exists(imageprocessor_views_path):
    with open(imageprocessor_views_path, 'w') as f:
        f.write("""
from django.http import JsonResponse

def process_image(request):
    return JsonResponse({"message": "Обработчик пока не настроен"})
""")
    print("Файл views.py для приложения imageprocessor создан.")
else:
    print("Файл views.py для приложения imageprocessor уже существует.")


def update_react_component_with_new_url(public_url):
    # Переходим в папку сервера react
    %cd /content/myproject/react-app

    # Извлечение URL с использованием регулярного выражения
    public_url_str = str(public_url)
    match = re.search(r'https://[^\s"]+', public_url_str)

    if match:
        link_r = match.group(0)
    else:
        print("URL не найден.")
        return

    # Создаем новый React-компонент с актуальным адресом
    text_ob_react = f"""
import React, {{ useState }} from "react";
import "./styles.css";
import {{ Grid, Box, Button, Typography, Card, Snackbar, Alert }} from "@mui/material";
import UploadFileIcon from "@mui/icons-material/UploadFile";
import SendIcon from "@mui/icons-material/Send";

function App() {{
  const [selectedFile, setSelectedFile] = useState(null);
  const [isProcessing, setIsProcessing] = useState(false);
  const [result, setResult] = useState(null);
  const [imageWithBoxes, setImageWithBoxes] = useState(null);
  const [snackbarOpen, setSnackbarOpen] = useState(false);
  const [logs, setLogs] = useState([]); // Логи обработки

  const handleFileChange = (event) => {{
    const file = event.target.files[0];
    setSelectedFile(file);
    setLogs([]); // Сброс логов
    setResult(null);
    setImageWithBoxes(null);
    if (file) {{
      setSnackbarOpen(true); // Показываем уведомление
    }}
  }};

  const handleSnackbarClose = () => {{
    setSnackbarOpen(false);
  }};

  const handleSubmit = async (e) => {{
    e.preventDefault();

    if (!selectedFile) {{
      alert("Пожалуйста, выберите изображение для загрузки");
      return;
    }}

    const formData = new FormData();
    formData.append("file", selectedFile);

    setIsProcessing(true);

    try {{
      const response = await fetch(`{link_r}/api/yolo/`, {{
        method: "POST",
        body: formData,
      }});

      if (!response.ok) {{
        throw new Error("Ошибка сервера");
      }}

      const data = await response.json();

      setLogs(data.detections);
      const format = selectedFile.type.includes("png") ? "png" : "jpeg";
      setImageWithBoxes(`data:image/${{format}};base64,${{data.image_data}}`);
    }} catch (error) {{
      console.error("Ошибка при обработке изображения:", error);
      alert("Произошла ошибка при распознавании. Попробуйте снова.");
    }} finally {{
      setIsProcessing(false);
    }}
  }};

  return (
    <Box
      sx={{{{
        background: "linear-gradient(180deg, #e3f2fd, #bbdefb)",
        height: "90vh",
        padding: "20px",
        display: "flex",
        flexDirection: "column",
      }}}}
    >
      {{/* Верхний блок: Изображение и Логи */}}
      <Grid container spacing={{2}} sx={{{{ flex: "1", marginBottom: "20px" }}}}>
        <Grid item xs={{6}} md={{6}}>
          <Card
            sx={{{{
              height: "95%",
              padding: "20px",
              boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
              textAlign: "right",
              backgroundColor: "#ffffff",
              borderRadius: "12px",
              position: "relative", // Для абсолютного позиционирования элементов внутри
            }}}}
          >
            <Typography variant="h5" sx={{{{ marginBottom: "20px" }}}}>
              Распознанное изображение
            </Typography>
            {{imageWithBoxes ? (
              <img
                src={{imageWithBoxes}}
                alt="Результат с рамками"
                style={{{{ maxWidth: "100%", height: "auto%", position: 'absolute', top: 80, right: 20, left: 20, maxWidth: "calc(100% - 40px)" }}}}
              />
            ) : (
              <Typography variant="body1">отсутствует</Typography>
            )}}
          </Card>
        </Grid>

        <Grid item xs={{6}} md={{6}}>
          <Card
            sx={{{{
              height: "95%",
              padding: "20px",
              boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
              backgroundColor: "#ffffff",
              borderRadius: "12px",
              overflowY: "auto",
            }}}}
          >
            <Typography variant="h5" sx={{{{ marginBottom: "20px" }}}}>
              Результаты распознавания
            </Typography>
            {{logs.length > 0 ? (
              <ul>
                {{logs.map((log, index) => (
                  <li key={{index}}>
                    {{`Класс: ('${{log.class}}'), вероятность: ${{log.confidence.toFixed(2)}}`}}
                  </li>
                ))}}
              </ul>
            ) : (
              <Typography variant="body1">нет данных</Typography>
            )}}
          </Card>
        </Grid>
      </Grid>

      {{/* Нижний блок: Управление */}}
      <Box
        sx={{{{
          padding: "10px",
          backgroundColor: "#ffffff",
          borderRadius: "12px",
          boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
          display: "flex",
          justifyContent: "space-between",
          alignItems: "center",
        }}}}
      >
        <Button
          variant="contained"
          component="label"
          color="primary"
          startIcon={{<UploadFileIcon />}}
          sx={{{{ width: "45%" }}}}
        >
          Загрузить
          <input type="file" hidden accept="image/*" onChange={{handleFileChange}} />
        </Button>
        <Button
          variant="contained"
          color="secondary"
          onClick={{handleSubmit}}
          disabled={{isProcessing}}
          endIcon={{<SendIcon />}}
          sx={{{{ width: "45%" }}}}
        >
          {{isProcessing ? "Обрабатываем..." : "Распознать"}}
        </Button>
      </Box>

      <Snackbar
        open={{snackbarOpen}}
        autoHideDuration={{3000}}
        onClose={{handleSnackbarClose}}
        anchorOrigin={{{{ vertical: "bottom", horizontal: "center" }}}}
      >
        <Alert onClose={{handleSnackbarClose}} severity="success" sx={{{{ width: "100%" }}}}>
          Изображение успешно загружено!
        </Alert>
      </Snackbar>
    </Box>
  );
}}

export default App;
"""

    # Сохраняем обновленный файл App.js
    with open("src/App.js", "w") as file:
        file.write(text_ob_react)
    print("React компонент успешно обновлен с новым адресом.")

# Переходим в папку сервера react
%cd /content/myproject/react-app

# Создаем динамическую ссылку
link_r = ""

text_ob_react = f"""
import React, {{ useState }} from "react";
import "./styles.css";
import {{ Grid, Box, Button, Typography, Card, Snackbar, Alert }} from "@mui/material";
import UploadFileIcon from "@mui/icons-material/UploadFile";
import SendIcon from "@mui/icons-material/Send";

function App() {{
  const [selectedFile, setSelectedFile] = useState(null);
  const [isProcessing, setIsProcessing] = useState(false);
  const [result, setResult] = useState(null);
  const [imageWithBoxes, setImageWithBoxes] = useState(null);
  const [snackbarOpen, setSnackbarOpen] = useState(false);
  const [logs, setLogs] = useState([]); // Логи обработки

  const handleFileChange = (event) => {{
    const file = event.target.files[0];
    setSelectedFile(file);
    setLogs([]); // Сброс логов
    setResult(null);
    setImageWithBoxes(null);
    if (file) {{
      setSnackbarOpen(true); // Показываем уведомление
    }}
  }};

  const handleSnackbarClose = () => {{
    setSnackbarOpen(false);
  }};

  const handleSubmit = async (e) => {{
    e.preventDefault();

    if (!selectedFile) {{
      alert("Пожалуйста, выберите изображение для загрузки");
      return;
    }}

    const formData = new FormData();
    formData.append("file", selectedFile);

    setIsProcessing(true);

    try {{
      const response = await fetch(`{link_r}/api/yolo/`, {{
        method: "POST",
        body: formData,
      }});

      if (!response.ok) {{
        throw new Error("Ошибка сервера");
      }}

      const data = await response.json();

      setLogs(data.detections);
      const format = selectedFile.type.includes("png") ? "png" : "jpeg";
      setImageWithBoxes(`data:image/${{format}};base64,${{data.image_data}}`);
    }} catch (error) {{
      console.error("Ошибка при обработке изображения:", error);
      alert("Произошла ошибка при распознавании. Попробуйте снова.");
    }} finally {{
      setIsProcessing(false);
    }}
  }};

  return (
    <Box
      sx={{{{
        background: "linear-gradient(180deg, #e3f2fd, #bbdefb)",
        height: "90vh",
        padding: "20px",
        display: "flex",
        flexDirection: "column",
      }}}}
    >
      {{/* Верхний блок: Изображение и Логи */}}
      <Grid container spacing={{2}} sx={{{{ flex: "1", marginBottom: "20px" }}}}>
        <Grid item xs={{6}} md={{6}}>
          <Card
            sx={{{{
              height: "95%",
              padding: "20px",
              boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
              textAlign: "right",
              backgroundColor: "#ffffff",
              borderRadius: "12px",
              position: "relative", // Для абсолютного позиционирования элементов внутри
            }}}}
          >
            <Typography variant="h5" sx={{{{ marginBottom: "20px" }}}}>
              Распознанное изображение
            </Typography>
            {{imageWithBoxes ? (
              <img
                src={{imageWithBoxes}}
                alt="Результат с рамками"
                style={{{{ maxWidth: "100%", height: "auto%", position: 'absolute', top: 80, right: 20, left: 20, maxWidth: "calc(100% - 40px)" }}}}
              />
            ) : (
              <Typography variant="body1">отсутствует</Typography>
            )}}
          </Card>
        </Grid>

        <Grid item xs={{6}} md={{6}}>
          <Card
            sx={{{{
              height: "95%",
              padding: "20px",
              boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
              backgroundColor: "#ffffff",
              borderRadius: "12px",
              overflowY: "auto",
            }}}}
          >
            <Typography variant="h5" sx={{{{ marginBottom: "20px" }}}}>
              Результаты распознавания
            </Typography>
            {{logs.length > 0 ? (
              <ul>
                {{logs.map((log, index) => (
                  <li key={{index}}>
                    {{`Класс: ('${{log.class}}'), вероятность: ${{log.confidence.toFixed(2)}}`}}
                  </li>
                ))}}
              </ul>
            ) : (
              <Typography variant="body1">нет данных</Typography>
            )}}
          </Card>
        </Grid>
      </Grid>

      {{/* Нижний блок: Управление */}}
      <Box
        sx={{{{
          padding: "10px",
          backgroundColor: "#ffffff",
          borderRadius: "12px",
          boxShadow: "0px 4px 12px rgba(0, 0, 0, 0.1)",
          display: "flex",
          justifyContent: "space-between",
          alignItems: "center",
        }}}}
      >
        <Button
          variant="contained"
          component="label"
          color="primary"
          startIcon={{<UploadFileIcon />}}
          sx={{{{ width: "45%" }}}}
        >
          Загрузить
          <input type="file" hidden accept="image/*" onChange={{handleFileChange}} />
        </Button>
        <Button
          variant="contained"
          color="secondary"
          onClick={{handleSubmit}}
          disabled={{isProcessing}}
          endIcon={{<SendIcon />}}
          sx={{{{ width: "45%" }}}}
        >
          {{isProcessing ? "Обрабатываем..." : "Распознать"}}
        </Button>
      </Box>

      <Snackbar
        open={{snackbarOpen}}
        autoHideDuration={{3000}}
        onClose={{handleSnackbarClose}}
        anchorOrigin={{{{ vertical: "bottom", horizontal: "center" }}}}
      >
        <Alert onClose={{handleSnackbarClose}} severity="success" sx={{{{ width: "100%" }}}}>
          Изображение успешно загружено!
        </Alert>
      </Snackbar>
    </Box>
  );
}}

export default App;
"""

# Сохраняем файл App.js
with open("src/App.js", "w") as file:
    file.write(text_ob_react)


text_ob_styles = """
body {
  margin: 0;
  font-family: "Roboto", sans-serif;
  background: linear-gradient(180deg, #e3f2fd, #bbdefb);
  min-height: 100vh;
}
"""
# Сохраняем файл styles.css
with open("src/styles.css", "w") as file:
    file.write(text_ob_styles)

import os
import json
import cv2
import numpy as np
import logging
from django.http import JsonResponse
from django.views.decorators.csrf import csrf_exempt
from ultralytics import YOLO
from io import BytesIO
from django.core.files.base import ContentFile
import base64
import time  # Для измерения времени

# Настраиваем логирование
logger = logging.getLogger('views')

# Загружаем YOLOv8 модель
try:
    model = YOLO('imageprocessor/models/yolov8n.pt')
except Exception as e:
    logger.error(f"Ошибка при загрузке модели YOLOv8: {str(e)}")

text_ob_django = """
import os
import json
import cv2
import numpy as np
import logging
from django.http import JsonResponse
from django.views.decorators.csrf import csrf_exempt
from ultralytics import YOLO
from io import BytesIO
from django.core.files.base import ContentFile
import base64
import time  # Для измерения времени

# Настраиваем логирование
logger = logging.getLogger('views')

# Загружаем YOLOv8 модель
try:
    model = YOLO('imageprocessor/models/yolov8n.pt')
except Exception as e:
    logger.error(f"Ошибка при загрузке модели YOLOv8: {str(e)}")

@csrf_exempt
def process_image(request):
    if request.method == 'POST' and request.FILES.get('file'):
        try:
            # Проверяем наличие директории и создаём её, если необходимо
            upload_dir = '/content/myproject/imageprocessor/uploads/'
            os.makedirs(upload_dir, exist_ok=True)

            # Определяем путь для временного файла
            uploaded_file = request.FILES['file']
            file_name = uploaded_file.name
            file_extension = os.path.splitext(file_name)[-1].lower()  # Получаем расширение файла
            file_path = os.path.join(upload_dir, f"temp_image{file_extension}")

            # Сохраняем загруженное изображение
            with open(file_path, 'wb') as f:
                for chunk in uploaded_file.chunks():
                    f.write(chunk)

            logger.info(f"Модель yolov8n начала распознавание файла '{file_name}'.")

            # Начало таймера
            start_time = time.time()

            # Выполняем распознавание
            results = model(file_path)

            # Конец таймера
            end_time = time.time()
            duration = end_time - start_time
            logger.info(f"Распознавание завершено за {duration:.2f} секунд.")

            # Загружаем изображение для обработки
            img = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)  # Поддержка прозрачности PNG

            # Преобразуем результаты и рисуем рамки
            detections = []
            for result in results:
                for box in result.boxes:
                    cls = int(box.cls[0])
                    conf = float(box.conf[0])
                    coords = box.xyxy[0].tolist()
                    detection = {
                        'class': model.names[cls],
                        'confidence': conf,
                        'coordinates': coords,
                    }
                    detections.append(detection)

                    # Логируем идентификацию
                    logger.info(f"Класс: {model.names[cls]}, "
                                f"вероятность: {conf:.2f} ")

                    # Рисуем рамки
                    x1, y1, x2, y2 = map(int, coords)
                    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
                    cv2.putText(img, f"{model.names[cls]}: {conf:.2f}",
                                (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

            # Кодируем изображение в base64
            output_extension = '.png' if file_extension == '.png' else '.jpg'
            _, img_encoded = cv2.imencode(output_extension, img)
            img_base64 = base64.b64encode(img_encoded).decode('utf-8')

            # Возвращаем JSON с результатами и изображением в base64
            return JsonResponse({
                'detections': detections,
                'image_data': img_base64
            })

        except Exception as e:
            logger.error(f"Ошибка обработки: {str(e)}")
            return JsonResponse({'error': str(e)}, status=500)

    logger.warning("Некорректный запрос")
    return JsonResponse({'error': 'Invalid request or no file provided'}, status=400)
"""

# Записываем код в файл Django view
views_file_path = '/content/myproject/imageprocessor/views.py'
with open(views_file_path, 'w') as views_file:
    views_file.write(text_ob_django)
```

```python
# Переход в директорию приложения imageprocessor
%cd /content/myproject/imageprocessor

# Создание директории models, если она еще не существует
!mkdir models

# Переход в директорию models
%cd /content/myproject/imageprocessor/models

# Загрузка YOLOv8 модели (yolov8n.pt) с официального репозитория Ultralytics
# Опция -nc (no-clobber) предотвращает повторную загрузку, если файл уже существует
!wget -nc https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8n.pt

# Импортирование и загрузка YOLOv8 модели в Python
try:
    # Указываем путь к модели, чтобы загрузить её в память
    model = YOLO('imageprocessor/models/yolov8n.pt')
except Exception as e:
    # Логируем ошибку, если модель не загрузилась
    logger.error(f"Ошибка при загрузке модели YOLOv8: {str(e)}")

@csrf_exempt
def process_image(request):
    if request.method == 'POST' and request.FILES.get('file'):
        try:
            # Проверяем наличие директории и создаём её, если необходимо
            upload_dir = '/content/myproject/imageprocessor/uploads/'
            os.makedirs(upload_dir, exist_ok=True)

            # Определяем путь для временного файла
            uploaded_file = request.FILES['file']
            file_name = uploaded_file.name
            file_extension = os.path.splitext(file_name)[-1].lower()  # Получаем расширение файла
            file_path = os.path.join(upload_dir, f"temp_image{file_extension}")

            # Сохраняем загруженное изображение
            with open(file_path, 'wb') as f:
                for chunk in uploaded_file.chunks():
                    f.write(chunk)

            logger.info(f"Модель yolov8n начала распознавание файла '{file_name}'.")

            # Начало таймера
            start_time = time.time()

            # Выполняем распознавание
            results = model(file_path)

            # Конец таймера
            end_time = time.time()
            duration = end_time - start_time
            logger.info(f"Распознавание завершено за {duration:.2f} секунд.")

            # Загружаем изображение для обработки
            img = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)

            # Преобразуем результаты и рисуем рамки
            detections = []
            for result in results:
                for box in result.boxes:
                    cls = int(box.cls[0])
                    conf = float(box.conf[0])
                    coords = box.xyxy[0].tolist()
                    detection = {
                        'class': model.names[cls],
                        'confidence': conf,
                        'coordinates': coords,
                    }
                    detections.append(detection)

                    # Логируем идентификацию
                    logger.info(f"Класс: {model.names[cls]}, "
                                f"вероятность: {conf:.2f} ")

                    # Рисуем рамки
                    x1, y1, x2, y2 = map(int, coords)
                    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
                    cv2.putText(img, f"{model.names[cls]}: {conf:.2f}",
                                (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

            # Кодируем изображение в base64 для передачи в фронтэнд
            output_extension = '.png' if file_extension == '.png' else '.jpg'
            _, img_encoded = cv2.imencode(output_extension, img)
            img_base64 = base64.b64encode(img_encoded).decode('utf-8')

            # Возвращаем JSON с результатами и изображением в base64
            return JsonResponse({
                'detections': detections,
                'image_data': img_base64
            })

        except Exception as e:
            logger.error(f"Ошибка обработки: {str(e)}")
            return JsonResponse({'error': str(e)}, status=500)

    logger.warning("Некорректный запрос")
    return JsonResponse({'error': 'Invalid request or no file provided'}, status=400)
```
